#Outlier remove

```python
import numpy as np

def detect_outliers_z_score(data):
    outliers_indices = []
    threshold = 4
    mean = np.mean(data)
    std = np.std(data)
    for i, value in enumerate(data):
        z_score = (value - mean) / std
        if np.abs(z_score) > threshold:
            outliers_indices.append(i)
    return outliers_indices

# List to store outlier counts for each column
outlier_counts_all = {}

# Iterate over each numerical column in the dataframe
for column in df.select_dtypes(include=['float64']).columns:
    # Apply outlier detection to the column
    outlier_indices = detect_outliers_z_score(df[column])
    # Count the number of outliers for the column
    outlier_counts_all[column] = len(outlier_indices)

# Print summary of outlier counts for each column
for column, count in outlier_counts_all.items():
    print("Number of outliers in column '{}': {}".format(column, count))
```

```python
# Create a copy of the DataFrame
df_cleaned = df.copy()

# Reset index to ensure alignment with outlier indices
df_cleaned.reset_index(drop=True, inplace=True)

# Iterate over each numerical column in the DataFrame
for column in df.select_dtypes(include=['float64']).columns:
    # Apply outlier detection to the column
    outlier_indices = detect_outliers_z_score(df[column])
    # Filter outlier indices within the range of DataFrame indices
    outlier_indices = [index for index in outlier_indices if index < len(df_cleaned)]
    # Remove rows with outlier indices
    df_cleaned = df_cleaned.drop(df_cleaned.index[outlier_indices])

# Print the shape of the cleaned DataFrame
print("Shape of cleaned DataFrame:", df_cleaned.shape)
```

Correlation Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr, kstest, norm
import numpy as np

# Load your DataFrame
# df_cleaned = pd.read_csv('path_to_your_dataframe.csv')  # Uncomment if loading from a CSV file

# Initial setup
data_point_color = '#40679E'
features = df_cleaned.columns.drop('Diabetes_binary')  # Exclude the target variable from the features

# Set up the plot
fig, ax = plt.subplots(len(features), 1, figsize=(10, 5 * len(features)))

# Analyze each feature
for i, feature in enumerate(features):
    # Check distribution of the feature using Kolmogorov-Smirnov test against a normal distribution
    data = df_cleaned[feature].dropna()  # Drop NA values for the test
    _, p_value_normal = kstest(data, 'norm', args=(data.mean(), data.std()))

    # Decide which correlation coefficient to use based on the K-S test result
    if p_value_normal > 0.05:
        # Data follows a normal distribution
        correlation_coefficient, p_value = pearsonr(df_cleaned['Diabetes_binary'], df_cleaned[feature])
        correlation_type = "Pearson"
    else:
        # Data does not follow a normal distribution
        correlation_coefficient, p_value = spearmanr(df_cleaned['Diabetes_binary'], df_cleaned[feature])
        correlation_type = "Spearman"

    # Create scatter plot with regression line
    sns.regplot(x=feature, y='Diabetes_binary', data=df_cleaned, ax=ax[i], line_kws={'color': 'red'}, scatter_kws={'color': data_point_color})

    # Annotate plot with correlation coefficient and p-value
    ax[i].text(0.05, 0.95, f"{correlation_type} score: {correlation_coefficient:.2f}\nP-value: {p_value:.2e}",
               horizontalalignment='left', verticalalignment='top', transform=ax[i].transAxes)

    # Set labels
    ax[i].set_ylabel('Diabetes Presence')
    ax[i].set_title(f'Relationship between {feature} and Diabetes Presence')

    # Save the current plot
    plt.savefig(f'plot_{feature}.png')  # Save each plot as a PNG file

# Adjust layout
plt.tight_layout()

# Display all plots
plt.show()
```

U test

```python
import pandas as pd
from scipy.stats import mannwhitneyu, ttest_ind, kstest, norm
import numpy as np

# Assuming 'df_cleaned' is your DataFrame already loaded with the necessary data

# Filter the DataFrame into two groups based on the 'Diabetes_binary' column
group_0 = df_cleaned[df_cleaned['Diabetes_binary'] == 0]  # Non-diabetic group
group_1 = df_cleaned[df_cleaned['Diabetes_binary'] == 1]  # Diabetic group

# List of numeric features to test
features = df_cleaned.select_dtypes(include=[np.number]).columns.drop('Diabetes_binary')

# Prepare the results list
results = []

# Perform statistical tests for each feature
for feature in features:
    # Calculate mean and standard deviation for both groups
    mean_0 = group_0[feature].mean()
    sd_0 = group_0[feature].std()
    mean_1 = group_1[feature].mean()
    sd_1 = group_1[feature].std()

    # Test for normality in each group using K-S test
    _, p_normal_0 = kstest(group_0[feature].dropna(), 'norm', args=(mean_0, sd_0))
    _, p_normal_1 = kstest(group_1[feature].dropna(), 'norm', args=(mean_1, sd_1))

    # Choose test based on normality of data
    if p_normal_0 < 0.05 or p_normal_1 < 0.05:
        # Data is not normal; use Mann-Whitney U Test
        test_stat, p_value = mannwhitneyu(group_0[feature], group_1[feature], alternative='two-sided')
        test_description = f'U={test_stat:.0f}'
    else:
        # Data is normal; use T-test
        test_stat, p_value = ttest_ind(group_0[feature], group_1[feature], equal_var=False)
        test_description = f't({test_stat:.2f})'

    # Append the results
    results.append({
        'Usage Data': feature,
        'Diabetics M': mean_1,
        'Diabetics SD': sd_1,
        'Non-Diabetics M': mean_0,
        'Non-Diabetics SD': sd_0,
        'Test Stat': test_description,
        'p value': p_value
    })

# Convert the results list to a DataFrame for better visualization
results_df = pd.DataFrame(results)
print(results_df)
```

Chi-square for validity test

```python
import pandas as pd
from scipy.stats import chi2_contingency
from sklearn.metrics import mean_squared_error
from math import sqrt


def validate_data_with_diabetes_binary(df, target='Diabetes_binary'):
    results = []
    for column in df.columns:
        if column != target:  # Skip the target column itself
            # Determine if the column is categorical or continuous by unique value count
            if df[column].nunique() <= 10:  # Assuming categorical variables have up to 10 unique values
                # Perform Chi-square test
                if df[column].nunique() > 1:  # Ensure there's at least two unique values for Chi-square
                    contingency_table = pd.crosstab(df[column], df[target])
                    chi2, p_value, _, _ = chi2_contingency(contingency_table)
                    results.append({'Variable': column,
                                    'Test Type': 'Chi-square',
                                    'Statistic': chi2,
                                    'p-value': p_value})
            else:
                # Calculate RMSE for continuous variables assuming predicted values are available
                predicted_col = f'predicted_{column}'
                if predicted_col in df.columns:
                    mse = mean_squared_error(df[column], df[predicted_col])
                    rmse = sqrt(mse)
                    results.append({'Variable': column,
                                    'Test Type': 'RMSE',
                                    'Statistic': rmse,
                                    'p-value': 'N/A'})

    return pd.DataFrame(results)

# Execute the validation function
validation_results = validate_data_with_diabetes_binary(df_cleaned)
print(validation_results)
```

```python
# The weighted average is taken in order to account for class imbalance
def measure_error(y_true, y_pred, label):
    return pd.Series({
        'balanced_accuracy': balanced_accuracy_score(y_true, y_pred),
        'precision': precision_score(y_true, y_pred, average='weighted'),
        'recall': recall_score(y_true, y_pred, average='weighted'),
        'f1': f1_score(y_true, y_pred, average='weighted'),
        'auc_roc': roc_auc_score(y_true, y_pred, average='weighted')
    }, name=label)


# Function to evaluate the model - fits the model, then computes training vs testing set metrics
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Computes the metrics for the training & testing set, and presents them in a dataframe
    train_test_error = pd.concat([measure_error(y_train, y_train_pred, 'Train'),
                                    measure_error(y_test, y_test_pred, 'Test')],
                                    axis=1)
    print(train_test_error, "\n")

    # Display classification report for test set
    print("Testing Set Classification Report:\n", classification_report(y_test, y_test_pred))

    # Display confusion matrix for the Test set
    cm = confusion_matrix(y_test, y_test_pred)
    disp = ConfusionMatrixDisplay(cm)

    return disp.plot()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
import xgboost as xgb

def evaluate_model(classifier, X_train, y_train, X_test, y_test):
    # Fit the classifier on the training data
    classifier.fit(X_train, y_train)

    # Predict the test data
    y_pred = classifier.predict(X_test)

    # Generate the confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Plotting the confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap='Reds', cbar=False)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')

    # Save the figure
    plt.savefig('confusion_matrix.png')  # Saves the confusion matrix to a file
    plt.close()  # Close the figure to free up memory

    # Print accuracy just for verification
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy Score: {accuracy:.2f}")

# Calculate the ratio of negative to positive class samples
neg_class_samples = len(y_train[y_train == 0])
pos_class_samples = len(y_train[y_train == 1])
scale_pos_weight = neg_class_samples / pos_class_samples

# Initialize the XGBoost classifier with predefined optimal parameters
xgb_classifier = xgb.XGBClassifier(
    booster='gbtree',
    scale_pos_weight=scale_pos_weight,
    learning_rate=0.4,
    max_depth=5,
    subsample=0.9,
    random_state=88
)

# Evaluate the model
evaluate_model(xgb_classifier, X_train, y_train, X_test, y_test)
```
